```python
class Stack():

    def __init__(self):
        self.elements = []

    def isEmpty(self):
        return self.elements == []

    def push(self,element):
        self.elements.append(element)

    def pop(self):
        self.elements.pop()

    def showLast(self):
        return self.elements[len(self.elements)-1] # veya elements[-1]

    def size(self):
        return len(self.elements)
```

```python
myStack = Stack()
```

```python
myStack.isEmpty()
```

```python
myStack.push(10)
myStack.push(20)
myStack.push(40)
```

```python
myStack.showLast()
```
