```python
class Queue():

    def __init__(self):
        self.elements = []

    def isEmpty(self):
        return self.elements == []

    def enqueue(self,element):
        self.elements.insert(0,element) # sona değil başa eklemek istediğimiz için insert kullanarak 0 a ekle diyoruz.

    def dequeue(self):
        return self.elements.pop() # en başa sürekli ekleme yapıldığında ilk giren en sona gidecek ve ilk giren ilk çıkmış olacak

    def size(self):
        return len(self.elements)
```

```python
myQueue = Queue()
```

```python
myQueue.isEmpty()
```

```python
myQueue.enqueue(10)
myQueue.enqueue(20)
myQueue.enqueue(30)
```

```python
myQueue.dequeue()
```

```python
myQueue.size()
```
