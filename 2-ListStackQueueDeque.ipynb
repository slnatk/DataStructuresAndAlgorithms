```python
# CONTAINS DUPLICATE PROBLEM
```

```python
# sıralama algoritmalarında en ideal zaman -> O(nlogn)
```

```python
myList = [1,2,3,4,5,6,7,8,2]
```

```python
def solution():
    hashSet = set()
    for num in myList:
        if num in hashSet:
            return True
        hashSet.add(num)  
    return False

# bir hashset oluştur, bu sete duplicate değerler barınamaz, listedeki numaraları tek tek deneyip hashsete ekleniyor ancak
# eğer eklemeye çalıştığım numara zaten listede varsa True döndürüyor yoksa False döndürüyor.
```

```python
solution()
```

```python
# SINGLE NUMBER PROBLEM: given a non-empty array of integers nums, every element appears twice except for one.
# Find that signle one. You must implement a solution with a linear runtime compelxity and use only constant extra space
```

```python
# bit manipulation (xor) O(N) and O(1) -< hafızada ekstra bir yer açmıyoruz ve listenin üzerinden sadece bir kere geçiyoruz.
```

```python
def singleNumber():
    myList = [1,1,2,2,3]

    result = 0

    for num in myList:
        result = num ^ result # xor

    return result    
```

```python
singleNumber()
```

```python
# MAJORITY ELEMENT PROBLEM
# given an array nums of sze n, return the majority element. The majority element is the element that appears more than n/2 times.
#You may assume that the majority element always exists in the array
```

```python
myList = [2,3,4,1,1,1,1,1,5]
def findMajority():
    numbers = {}
    result = 0
    maxNumber = 0

    for num in myList:
        numbers[num] = 1 + numbers.get(num,0)
        if numbers[num] > maxNumber:
            result = num 
        maxNumber = max(maxNumber,numbers[num])   

    return result  
```

```python
findMajority() # O(N) tek bir for loop var 
```

```python
# Boyer moore algoritması

def boyerMoore():
    result = 0
    count = 0

    for num in myList:
        if count == 0:
            result = num
        count += 1 if num == result else -1
    return result    
    
```

```python
boyerMoore()  # ekstra alan kullanmadığımız için (yeni bir hashsetvs.) space -> O(1)  
               
```
