```python
'''
Given the head of a linked list, remove the nth node from the end of the list and return its head.

'''
```

```python

class Solution:
    def removeNthFromEnd(self, head: Optional[ListNode], n: int):
        
        leftPointer = head
        rightPointer = head
        
        while n > 0 and rightPointer: # iki pointer arasında 2 boşluk var.
            rightPointer = rightPointer.next
            n -= 1
        
        while rightPointer and rightPointer.next:  # rightPointer.next None ise dur
            leftPointer = leftPointer.next
            rightPointer = rightPointer.next
                
        if leftPointer == head and not rightPointer: #  eğer tek elemanlı bir liste verildiyse sonucu null döndür.
            return head.next 

        leftPointer.next = leftPointer.next.next
        
        return head

# Sliding Window tekniği
# Time -> O(N) , Space -> O(1)
```
