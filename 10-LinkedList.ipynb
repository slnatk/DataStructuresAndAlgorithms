```python
# Singly Linked List
```

```python
class Node():

    def __init__(self,value):

        self.value = value
        self.nextNode = None
        
```

```python
# Liste yapmak aslında şudur;
firstNode = Node(10)
```

```python
secondNode = Node(20)
```

```python
thirdNode = Node(30)
```

```python
firstNode.nextNode = secondNode # bu şekilde birbirine de bağlamış olduk
```

```python
secondNode.nextNode = thirdNode
```

```python
firstNode.nextNode.value # 2.nin değerine 1. üzerinden erişebiliriz
```

```python
firstNode.nextNode.nextNode.value # 3.ye bile erişebiliriz
```

```python
class DoublyNode():

    def __init__(self,value):
        self.value = value
        self.nextNode = None
        self.prevNode = None
        
```

```python
x = DoublyNode(5)
y = DoublyNode(10)
z = DoublyNode(15)
```

```python
x.nextNode = y
y.prevNode = x # birbirlerine bağlamış olduk.
```

```python
y.nextNode=z
z.prevNode=y
```

```python
x.nextNode.prevNode.value
```
