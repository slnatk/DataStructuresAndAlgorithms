```python
def bigon(n):
    for i in range(0,n):
        print(i)
```

```python
bigon(10)
```

```python
def bigon2(n):
    for i in range(0,n):
        for j in range(0,n):
            print(i,j)
```

```python
bigon2(5) # listenin içinde iki kere gezmiş olduk
```

```python
def bigon3(n):
    for i in range(0,n):
        for j in range(0,n):
            for k in range(0,n):
                print(i,j,k)
    
```

```python
bigon3(5)
```

```python
import math
# math.floor() -> yuvarlama

def logn(n):
    while n > 1:
        n = math.floor(n/2)
        print(n)
```

```python
logn(16)
```

```python
math.floor (12/5)
```

```python
def nlogn(n):
    lim = n
    while n >1:
        n = math.floor(n/2)
        for i in range (1,lim):
            print(i)
```

```python
nlogn(16)
```

```python
def nfactorial(n):
    if n == 0:
        print("1")
        return
    else:
        for i in range(0,n):
            print(i)
            nfactorial(n-1)  # recursive, örneğin 5 bittiğinde 4 için çağrılacak...
```

```python
nfactorial(5)
```

```python
counter = 0
def actualFactorial(n):
    global counter 
    counter += 1
    print(counter)
    if (n == 1 or n == 0):
        return 1
    else:
      return (n * actualFactorial(n-1))  
      
```

```python
actualFactorial(3) #  defa çağrıldı. yani complexity = O(n)
```
